> **Data Description in [Kaggle](https://www.kaggle.com/mkechinov/ecommerce-events-history-in-electronics-store)**    
>  * category_code: Product's category taxonomy (code name) if it was possible to make it. `Usually present for meaningful categories and skipped for different kinds of accessories.`     
>  * brand: Downcased string of brand name. `Can be missed.`
>  * user_session: Temporary user's session ID. Same for each user's session. Is changed every time user come back to online store from a long pause.

위의 데이터 설명에 따라, category_code와 brand는 있을 수도 있고, 없을 수도 있는 컬럼입니다. 하지만, user_session은 각 user_id마다 모두 가지고 있어야 하는 값이므로, 결측치가 발생한 경우를 제거해줘야합니다.     
이때, 제거 외에도, user_id와 event_time, event_type 등 다른 컬럼들을 고려하여 값을 추축하여 대체하는 방법도 있지만, `user_session의 결측치 비중이 약 0.2% 이므로 제거하는 방법을 택했습니다.`    
하지만, category_code나 brand 같은 경우 결측치가 전체 데이터의 약 30%를 차지합니다. 해당 경우에는 평균치, 0의 값 등으로 채우는 방법(Imputation)이 있을 수 있고, 회귀 방식을 활용하는 방법 등이 있으나, user기반의 분석을 진행할 예정이므로 일괄적으로 '0'으로 값을 대치하는 방식을 택했습니다.

> 피드백: 결측치 제거 기준 10%로 잡은 이유? 레퍼런스 있는지? 고객 데이터 10%라면 비중이 크지 않은가? 

> 일반적으로 10%의 결측치를 제거해준다는 점에서는 고객 군(클러스터) 데이터를 다루는데 있어서는 위험할 수 있습니다. 해당 데이터는 1% 미만의 결측치를 차지하고 있어 제거하는 방법을 택했습니다. 

> 블로그 글 인용: 개체를 제거하는 경우는 기준이 조금 다릅니다. 개체(표본)수에 비해 결측값을 가진 개체의 수나 비중이 확연히 적을 경우 제거를 고려해 볼 수도 있습니다. (이 기준은 보통 10% 미만 수준에서 논의가 이루어지고 있습니다.)
하지만, 클러스터링 분석에서 개체를 삭제하는 것은 좀더 고민이 필요합니다. 클러스터링 작업 이후에 결측값 때문에 제거한 개체도 결국 어느 시점에 이르러 클러스터를 할당해줘야하는 이슈가 발생할 수 있기 때문입니다. 예를 들어 클러스터링 분석은 고객세분화를 위한 목적으로 많이 활용되고 있기 때문에 결국 클러스터 모델링(결과도출)에 사용하지 않은 고객(정보)도 추후에 결측값을 보완하여 클러스터를 부여해야만 할 수도 있습니다. 좀더 쉽게 설명하자면, A라는 고객을 결측값 때문에 제외하고 0에서 4까지의 고객군(클러스터)을 나누었는데, 분석에 사용하지 않은 A라는 고객도 결국 나중에 (마케팅이나 다른 목적으로 사용하기 위해서) 0에서 4중 어느 한 가지에 할당해야만 하는 상황에 이를 수 있고, 이 과정에서 결국 결측값을 보완하는 작업이 필요할 수 있다는 의미입니다. 결국 조금 빠르던 늦던 간에 결측값을 다른 값으로 채워야 하는 상황에 맞닥뜨릴 수 있기 때문에 분석 목적까지 염두에 두고 결측값 처리 방법을 고려할 필요가 있습니다.
개체를 제거하는 것은 또한 선택의 문제로 귀결될 수도 있습니다. 클러스터 모델링 작업에 좀더 정확하고 확실한 데이터만 활용한다는 점에서 클러스터 분류 결과의 신뢰도를 높일 수도 있지만, 결측치를 미리 삭제하면서 데이터 편향(bias)이 발생할 가능성도 있습니다.

#### 일별 퍼널 단계별 전환율
* 전환: 고객이 마케팅에 반응하여 마케터가 의도한 행동을 하는 것
* view, cart, purchase를 각 퍼널로 고려하여 전환율 계산
* view => cart
* cart => purchase로의 전환이 있습니다. 

```python
raw_data.head()
```

```python
# 3가지 퍼널 확인
raw_data[(raw_data['event_type_cart'] == 0) & (raw_data['event_type_purchase'] == 0) & (raw_data['event_type_view'] == 0)]
```

```python
grouped = raw_data.groupby(['event_date'])
funnel_by_date = grouped[['event_type_view', 'event_type_cart', 'event_type_purchase']].sum().reset_index()
funnel_by_date
```

```python
funnel_by_date['event_sum'] = funnel_by_date['event_type_view'] + funnel_by_date['event_type_cart'] + funnel_by_date['event_type_purchase']
funnel_by_date
```

```python
# 전환율 계산
funnel_by_date['event_type_view'] = funnel_by_date['event_type_view'] / funnel_by_date['event_sum']*100
funnel_by_date['event_type_cart'] = funnel_by_date['event_type_cart'] / funnel_by_date['event_sum']*100
funnel_by_date['event_type_purchase'] = funnel_by_date['event_type_purchase'] / funnel_by_date['event_sum']*100
funnel_by_date
```

```python
del funnel_by_date['event_sum']
```

```python
# 시각화
plt.figure(figsize=(20, 50))
plt.barh(funnel_by_date['event_date'], funnel_by_date['event_type_view'])
plt.barh(funnel_by_date['event_date'], funnel_by_date['event_type_cart'], left = funnel_by_date['event_type_view'])
plt.barh(funnel_by_date['event_date'], funnel_by_date['event_type_purchase'], left = funnel_by_date['event_type_view']+funnel_by_date['event_type_cart'])
plt.title("일별 퍼널 단계별 전환율의 분포")
plt.xlabel(f"일자 {funnel_by_date['event_date'].loc[0]} ~ {funnel_by_date['event_date'].loc[len(funnel_by_date['event_date'])-1]}")
plt.xticks(color='w')
plt.ylabel("전환율[%]")
plt.grid(True)
plt.tight_layout()
plt.show()
```
