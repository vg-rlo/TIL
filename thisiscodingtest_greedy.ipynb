최대한 N을 많이 빼줘야 하는 문제이기 때문에 문제에서 주어진 순서대로 푸는 것이 아니라 먼저 N에 대해 많이 빼주는 과정을 먼저 수행하면 좋음.     
최소 횟수, 최소 비용 => 현상황에서 구해야하는 값에 유리한 조건을 선택 

```python
import sys
import io

sys.stdin = io.StringIO('1 2')
n, k = map(int, sys.stdin.readline().split())
```

```python
n , k
```

```python
def find_N(n, k):
    result = 0
    while True:
        target = (n//k) * k # 나누는 수에 대한 배수가 되도록  
        result += (n-target) # target을 빼서 n으로 나눠지는 수로 result 저장  
        n = target # n을 나누기 가능한 target으로 저장 
        if n < k: break # K보다 작으면 나눴을때 1이 나오지 않으므로 break 
#         if n== 1: break
        result += 1
        n //= k 

    result += (n-1) # break문 기준으로 안나눠질때는 1을 n-1개 만큼 빼준다. 
    print(result)
    return result
```

```python
n, k = map(int, input().split())

find_N(n, k)
```
