# 문제 9316 
양의 정수 n개가 주어졌을 때, 가능한 모든 쌍의 GCD의 합을 구하는 프로그램을 작성하시오.

* 입력: 첫째 줄에 테스트 케이스의 개수 t (1 ≤ t ≤ 100)이 주어진다. 각 테스트 케이스는 한 줄로 이루어져 있다. 각 테스트 케이스는 수의 개수 n (1 < n ≤ 100)가 주어지고, 다음에는 n개의 수가 주어진다. 입력으로 주어지는 수는 1,000,000을 넘지 않는다.
* 출력: 각 테스트 케이스마다 가능한 모든 쌍의 GCD의 합을 출력한다.

```python
def gcd(x, y):
    while y:
        x, y = y, x % y
    return x

print(gcd(48,60))
```

```python
i_num = int(input()) # input으로 몇개의 입력으로 테스트할지 결정 

for _ in range(i_num): # i_num 회차만큼 for문 반복 
    num_list = list(map(int, input().split())) # 공백을 기준으로 문자열을 int로 바꾸고 list에  mapping 
    iter_num = num_list[0] # 원소 간 비교 횟수
    num_list = num_list[1:]
    gcd_sum = 0
    for i in range(0, iter_num-1): # 마지막 수는 이미 앞선 수에서 다 비교해주니까 pass하기 때문에 -1 
        for j in range(i+1, iter_num): # 비교 기준 수 다음에 해당하는 수들끼리 비교 e.g. [0] - [1], [2], [3]
            gcd_sum += gcd(num_list[i], num_list[j]) # 두 수 간의 최대 공약수 계산
    print(gcd_sum)
```
