## Seaborn 특징

* 여러 변수 간의 관계를 검사하기 위한 데이터 집합 지향 API
* 범주형 변수를 사용하여 관측치 또는 집계 통계량을 표시하기 위한 전문적인 지원
* 일변량 또는 이변량 분포를 시각화하고 데이터의 부분 집합 간 비교하기 위한 옵션
* 서로 다른 종류의 종속 변수에 대한 선형 회귀 모형의 자동 추정 및 표시
* 복잡한 데이터셋의 전체 구조에 대한 편리한 보기
* 복잡한 시각화를 쉽게 구축할 수 있는 다중 플롯 그리드 구조를 위한 높은 수준의 추상화
* 여러 테마가 내장된 matplotlib 그림 스타일링 제어
* 데이터의 패턴을 충실히 나타내는 색상 팔레트 선택 도구

## Seaborn 양날의 칼 "kwargs" 
* 이런저런 옵션을 설정하게 해주는 기능
* kind에서 설정하는 plot의 종류에 따라 적절한 matplotlib 함수 또는 자체 함수를 호출해주기 때문에 그에 따른 keyword arguments를 넣어줄 수 있도록 설정되어 있음.
* 결론? 예를 들어, kind='line'으로 해준다면, matplotlib에서 지원하는 plot 종류라면 matplotlib의 문서에서 arguments들을 확인해줄 필요가 있고, 그에 따라 적절한 인자를 수정할때 참고해야한다는 점!

## 그래서 Matplotlib VS Seaborn 
* 초보자인데 시각화를 두둥탁!하고 해내고 싶다 => Seaborn 
* 세부적인 level 등의 조정이 필요하다? => Matplotlib 
* 시각화 쪽을 차곡차곡 배워나갈 것이다? => Matplotlib

```
import seaborn as sns
sns.__version__
sns.set(style='whitegrid') # 코드 스타일처럼 그래프 스타일 지정

# 함께 많이 쓰이는 라이브러리 
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## 산점도(Scatter Plot)

```
penguins = sns.load_dataset('penguins')
penguins
```

```
sns.relplot(x="bill_length_mm", y="bill_depth_mm", data=penguins) # default는 scatter plot
```

#### **아래 그래프에서 어떤 상관관계가 보이나?**

```
sns.relplot(x="flipper_length_mm", y="body_mass_g", data=penguins) # default는 scatter plot
```

```
# scatter plot에 hue에 수치형 데이터를 넣어주면 hue에 할당된 수치형 데이터를 구간화해서 그라데이션해준다. 
sns.relplot(x="bill_length_mm", y="bill_depth_mm", 
            hue="bill_length_mm", data=penguins) # default는 scatter plot
```

```
# scatter plot에 hue에 수치형 데이터를 넣어주면 hue에 할당된 수치형 데이터를 구간화해서 그라데이션해준다. 
# 이때 hue는 꼭 x, y 변수에 할당된 데이터일 필요가 없다. 
sns.relplot(x="bill_length_mm", y="bill_depth_mm", 
            hue="body_mass_g", data=penguins) # default는 scatter plot
```

```
# scatter plot에 hue에 수치형 데이터를 넣어주면 hue에 할당된 수치형 데이터를 구간화해서 그라데이해준다. 
sns.relplot(x="bill_length_mm", y="bill_depth_mm", 
            hue="body_mass_g", size="body_mass_g", data=penguins) # default는 scatter plot
```

```
# scatter plot에 hue에 수치형 데이터를 넣어주면 hue에 할당된 수치형 데이터를 구간화해서 그라데이해준다. 
sns.relplot(x="bill_length_mm", y="bill_depth_mm", 
            hue="body_mass_g", size="body_mass_g", sizes=(10, 300), 
            data=penguins) # default는 scatter plot
```

```
# style을 추가하면, 수치형 2개, 범주형 1개에 대해 표현 가능하다. 
sns.relplot(x="flipper_length_mm", y="body_mass_g",
            hue='species', style='species', data=penguins) # default는 scatter plot
```

```
# style을 추가하면, 수치형 2개, 범주형 2개에 대해 표현 가능하다. But, 복잡도 높아진다. 
sns.relplot(x="flipper_length_mm", y="body_mass_g",
            hue='species', style='island', data=penguins) # default는 scatter plot
```

```
# col을 추가하면, col에 해당하는 종(species)별로 창을 띄워준다. 
sns.relplot(x="flipper_length_mm", y="body_mass_g",
            hue='species', col='island', data=penguins) # default는 scatter plot
```

```
# col을 추가하면, col에 해당하는 종(species)별로 창을 띄워준다. 
sns.relplot(x="bill_length_mm", y="bill_depth_mm", 
            hue='species', col='island', data=penguins) # default는 scatter plot
```

```
dots = sns.load_dataset('dots')
dots
```

```
# 하늘색 영역은 실제값에 대한 95%의 신뢰구간을 의미합니다. 
# 이때의 default는 평균값을 기준으로 신뢰구간을 정합니다. 
sns.relplot(x='time', y='firing_rate', 
            kind='line', data=dots)
```

```
# 신뢰구간은 ci 옵션 변경
# 신뢰구간 표시 안 하고 싶을때는 None으로 설정 
sns.relplot(x='time', y='firing_rate', ci=None,
            kind='line', data=dots)
```

```
# 하늘색 영역은 실제값에 대한 95%의 신뢰구간을 의미합니다.  
sns.relplot(x='time', y='firing_rate', ci='sd', # 신뢰구간을 표준편차로 해주고 싶을때 사용
            kind='line', data=dots)
```

```
# 집계 데이터 기준이 아닌, 각 point들을 표현하고 싶을때 estimator 옵션 변경
sns.relplot(x='time', y='firing_rate', estimator=None,
            kind='line', data=dots)
```

```
# lineplot - hue 옵션 변경
# hue에 범주형 데이터 넣어주면 그에 따른 라벨링이 이뤄진다. 
sns.relplot(x='time', y='firing_rate', hue='choice',
            kind='line', data=dots)
```

```
# lineplot - hue 옵션 변경
# hue에 범주형 데이터 넣어주면 그에 따른 라벨링이 이뤄진다. 
sns.relplot(x='time', y='firing_rate', hue='align',
            kind='line', data=dots)
```

```
# lineplot - style 옵션 추가 
# hue에 범주형 데이터 넣어주면 그에 따른 라벨링이 이뤄진다. 
sns.relplot(x='time', y='firing_rate', hue='align', style='choice',
            kind='line', data=dots)
```

```
# lineplot - dashes, markers 옵션 변경
sns.relplot(x='time', y='firing_rate', hue='align', style='choice',
            kind='line', data=dots,
            dashes=False, markers=True)
```

```
# lineplot - col 옵션 추가 
sns.relplot(x='time', y='firing_rate', hue='align', style='choice',
            kind='line', data=dots, col='choice',
            dashes=False, markers=True)
```

```
# data에서 특정 data만 필터링해서 시각화 
# align이라는 범주형 데이터에서 sacc만 시각화 
sns.relplot(x='time', y='firing_rate', hue='align', style='choice',
            kind='line', data=dots.query("align=='sacc'"), col='choice',
            dashes=False, markers=True)
```

```
# data에서 특정 data만 필터링해서 시각화 
# align이라는 범주형 데이터에서 sacc만 시각화 
sns.relplot(x='time', y='firing_rate', hue='align', style='choice',
            kind='line', data=dots.query("align=='sacc'"), 
            dashes=False, markers=True)
```

```
# data에서 특정 data만 필터링해서 시각화 - hue에 수치형 데이터로 지정
# align이라는 범주형 데이터에서 sacc만 시각화 
sns.relplot(x='time', y='firing_rate', hue='coherence', style='choice',
            kind='line', data=dots.query("align=='sacc'"), 
            dashes=False, markers=True)
```

```
# 따라서 col으로 구분해주는 것이다.
sns.relplot(x='timepoint', y='signal', 
            col='region', size='event', hue='subject', 
            kind='line', data=fmri)
```

```
# 또는 seaborn의 palette 기능을 활용한다. 
palette = sns.cubehelix_palette(n_colors=14, light=0.8)
sns.relplot(x='timepoint', y='signal', 
            col='region', style='event', hue='subject', # 위의 경우 size 보다 style 구분이 더 시각화로 보기 좋아보임
            kind='line', data=fmri,
            palette=palette) # palette 옵션 추가 
```

```
# data에서 특정 부분만 시각화 - data.query 
palette = sns.cubehelix_palette(n_colors=14, light=0.8)
sns.relplot(x='timepoint', y='signal', 
            col='region',  hue='subject', # event에서 cue만 고려하므로 style 지정한 부분 삭제한다. 
            kind='line', data=fmri.query("event=='cue'"),
            palette=palette) 
```

```
# row 옵션 변경
sns.relplot(x='timepoint', y='signal', 
            row='event', col='region',  hue='subject',
            palette=palette, kind='line', data=fmri) 
```

```
# subject가 여러개라, 각 subject별로 살펴보는 것이 좋다. 
sns.relplot(x='timepoint', y='signal', 
            col='subject', col_wrap=4,  hue='event', style='event',
            kind='line', linewidth=3, data=fmri.query("region=='parietal'")) 
```

## 범주형 데이터(Categorical Data)

### 범주형 산점도(Categorical scatterplots)

* `stripplot()` (with `kind="strip"`; the default)
* `swarmplot()` (with `kind="swarm"`)

```
import seaborn as sns
penguins = sns.load_dataset('penguins')

penguins
```

```
sns.catplot(x='species', y='body_mass_g', data=penguins) # jittering 포함
```

```
sns.catplot(x='species', y='body_mass_g', 
            jitter=False, 
            data=penguins)
```

```
# 데이터간 겹치지 않도록 펼치기 - swarm
sns.catplot(x='species', y='body_mass_g', jitter=False,
            kind='swarm', # kind를 변경해줄때는 해당 plot이 지원하는 keyword arguments를 확인할 것 
            data=penguins)
```

```
# 데이터간 겹치지 않도록 펼치기 - swarm
sns.catplot(x='species', y='body_mass_g', 
            kind='swarm', # kind를 변경해줄때는 해당 plot이 지원하는 keyword arguments를 확인할 것 
            data=penguins)
```

```
# 데이터간 겹치지 않도록 펼치기 - swarm + hue
sns.catplot(x='species', y='body_mass_g', 
            hue='sex',
            kind='swarm', # kind를 변경해줄때는 해당 plot이 지원하는 keyword arguments를 확인할 것 
            data=penguins)
```

```
# 데이터간 겹치지 않도록 펼치기 - swarm
sns.catplot(x='sex', y='body_mass_g', 
            hue='species',
            kind='swarm', # kind를 변경해줄때는 해당 plot이 지원하는 keyword arguments를 확인할 것 
            data=penguins)
```

```
# 데이터간 겹치지 않도록 펼치기 - swarm + hue + order(순서 바꾸기)
sns.catplot(x='sex', y='body_mass_g', 
            hue='species',
            kind='swarm', # kind를 변경해줄때 는 해당 plot이 지원하는 keyword arguments를 확인할 것 
            order=['Female', 'Male'], data=penguins)
```

```
# 데이터간 겹치지 않도록 펼치기 - swarm + hue + x와 y축 데이터값 바꾸기
sns.catplot(x='body_mass_g', y='species', 
            hue='island',
            kind='swarm', # kind를 변경해줄때는 해당 plot이 지원하는 keyword arguments를 확인할 것 
            data=penguins)
```

```
# 데이터간 겹치지 않도록 펼치기 - swarm + hue + x와 y축 데이터값 바꾸기
sns.catplot(x='species', y='body_mass_g', 
            hue='sex', col='island', 
            kind='swarm', # kind를 변경해줄때는 해당 plot이 지원하는 keyword arguments를 확인할 것 
            data=penguins)
```

```
# 데이터간 겹치지 않도록 펼치기 - swarm + hue + x와 y축 데이터값 바꾸기 + 간격 줄이기/넓히기
sns.catplot(x='species', y='body_mass_g', 
            hue='sex', col='island', 
            aspect=.7, # 간격 줄이기 
            kind='swarm', # kind를 변경해줄때는 해당 plot이 지원하는 keyword arguments를 확인할 것 
            data=penguins)
```

#### 바이올린 플롯(Violin plots)

* `violinplot`: 커널 밀도 추정과 상자 도표 결합

```
# violin plot 
# 굴곡은 커널 밀도, 안에 작은 진한 막대는 box plot
sns.catplot(x='species', y='body_mass_g', 
            hue='sex', kind='violin', data=penguins)
```

```
# violin plot + bw, cut
# 굴곡은 커널 밀도, 안에 작은 진한 막대는 box plot
sns.catplot(x='species', y='body_mass_g', 
            hue='sex', kind='violin', 
            bw=.15, cut=0, data=penguins)
```

```
# violin plot, bw, cut + split 
# 굴곡은 커널 밀도, 안에 작은 진한 막대는 box plot
sns.catplot(x='species', y='body_mass_g', 
            hue='sex', kind='violin', 
            split=True, data=penguins)
```

```
# violin plot, bw, cut, split + box 대신 stick  
# 굴곡은 커널 밀도, 안에 작은 진한 막대는 box plot
sns.catplot(x='species', y='body_mass_g', 
            hue='sex', kind='violin', 
            inner='stick', 
            split=True, data=penguins)
```

```
# bar plot 
sns.catplot(kind='bar',
            x='origin', y='mpg',
            hue='cylinders',
            data=mpg)
```

```
# bar plot 
sns.catplot(kind='bar',
            x='origin', y='horsepower', # 마력을  y로 bar plot            
            hue='cylinders',
            palette='ch:.20', # 20개의 색으로 구분 
            data=mpg)
# 결과적으로 실린더 수에 따라 마력을 얼마나 가지는지 확인 가능
```

```
# bar plot 
sns.catplot(kind='bar',
            x='cylinders', y='horsepower', # 마력을  y로 bar plot            
            # hue='', # hue 제외 
            palette='ch:.20', # 20개의 색으로 구분 
            data=mpg)
# 검정색 중앙 막대는 오류를 얼만큼 포함하고 있는지 보여주는 것이다. 
```

```
# bar plot + edgecolor 
sns.catplot(kind='bar',
            x='cylinders', y='horsepower', # 마력을  y로 bar plot            
            # hue='', # hue 제외 
            edgecolor='.6', # edge 처리로 구분 확실해짐
            palette='ch:.20', 
            data=mpg)
```

#### 포인트 플롯(Point plots)

```
# point plot + x값과  hue값 교환
sns.catplot(kind='point',
            x='class', y='survived', 
            hue='who', 
            data=titanic)
# 시각화 결과 상대적으로 1등석 사람들이 많이 생존했음. 
# 3등석 사람들은 1, 2등서겡 비해 사망률이 높음
```

```
sns.catplot(kind='point',
            x='embark_town', y='survived',
            hue='who',
            data=titanic)
# 어디 선착장에서 탔는지, 시각화 결과에서 관계성이 크진 않아보임임
```

## 분포 시각화(Distribution Visualization)

### 일변량 분포(Univariate distributions)

```
import numpy as np
x = np.random.randn(200)
sns.distplot(x)
```

#### 히스토그램(Histograms)

```
# hitogram + kde, rug
sns.distplot(x, kde=False, rug=True) # kde=False, rug=True로 된 distplot은 히스토그램
```

```
# hitogram + bins
sns.distplot(x, kde=False, rug=True, bins=20) # kde=False, rug=True로 된 distplot은 히스토그램
```

#### 커널 밀도 추정(Kernel density estimation)

```
# histogram 없애고, 밀도만 보기 
sns.distplot(x, hist=False, rug=True)
```

```
sns.kdeplot(x, shade=True)
```

```
import seaborn as sns
sns.kdeplot(x)
sns.kdeplot(x, bw=.2, label='bw: 0.2')
sns.kdeplot(x, bw=1, label='bw: 1')

import matplotlib.pyplot as plt
plt.legend()
# 강의 결과와 다르게 나옴.. ? 해결해보자 
```

```
# kdeplot + rugplot, 
sns.kdeplot(x, shade=True, cut=0) # cut=0이면 극단값 제외(그래프 양 끝 확인)
sns.rugplot(x)
```
